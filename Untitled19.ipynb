```python
import os
from rknn.api import RKNN

ONNX_MODEL = '/content/yolov5/runs/train/exp/weights/your_custom_yolov5_model.onnx'
RKNN_MODEL = 'yolov5s.rknn'
DATASET = './dataset'

QUANTIZE_ON = True

if __name__ == '__main__':
    # Create RKNN object
    rknn = RKNN()

    if not os.path.exists(ONNX_MODEL):
        print('Model does not exist')
        exit(-1)

    # Configure model preprocessing
    print('--> Config model')
    rknn.config(reorder_channel='0 1 2',
                mean_values=[[0, 0, 0]],
                std_values=[[255, 255, 255]],
                optimization_level=3,
                target_platform='rv1126',
                output_optimize=1,
                quantized_dtype='asymm8',  # You may need to adjust this based on your hardware
                quantize_input_node=QUANTIZE_ON)
    print('Configured model')

    # Load ONNX model
    print('--> Loading model')
    ret = rknn.load_onnx(model=ONNX_MODEL)
    if ret != 0:
        print('Load YOLOv5 model failed!')
        exit(ret)
    print('Loaded model')

    # Build model
    print('--> Building model')
    ret = rknn.build(target='rv1126', device_id='0', debug=True)
    if ret != 0:
        print('Build model failed!')
        exit(ret)
    print('Built model')

    # Export RKNN model
    print('--> Exporting RKNN model')
    ret = rknn.export_rknn(RKNN_MODEL)
    if ret != 0:
        print('Export RKNN model failed!')
        exit(ret)
    print('Exported RKNN model')

    rknn.release()
```
